```python
1. Write a program to accept percentage from the user and display the grade according to the following criteria:

Grade                                                     Marks

A                                                          >90
B                                                          >80 and <=90
C                                                          >=60 and <=80
D                                                           below 60

def calculate_grade(percentage):
    if percentage > 90:
        return 'A'
    elif percentage > 80:
        return 'B'
    elif percentage >= 60:
        return 'C'
    else:
        return 'D'

def main():
    try:
        percentage = float(input("Enter percentage: "))
        if percentage < 0 or percentage > 100:
            raise ValueError("Percentage should be between 0 and 100")
        
        grade = calculate_grade(percentage)
        print(f"Grade: {grade}")
    except ValueError as ve:
        print(f"Error: {ve}")

if __name__ == "__main__":
    main()

```

```python
2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:

Tax                           Cost Price(in Rs)
15%                             >100000
10%                      >50000 and <= 100000
5%                              <= 50000


def calculate_road_tax(cost_price):
    if cost_price > 100000:
        return cost_price * 0.15
    elif cost_price > 50000:
        return cost_price * 0.10
    else:
        return cost_price * 0.05

def main():
    try:
        cost_price = float(input("Enter the cost price of the bike (in Rs): "))
        if cost_price < 0:
            raise ValueError("Cost price should be a positive number")

        road_tax = calculate_road_tax(cost_price)
        print(f"Road tax to be paid: Rs {road_tax:.2f}")
    except ValueError as ve:
        print(f"Error: {ve}")

if __name__ == "__main__":
    main()
```

```python
3. Accept any city from the user and display monuments of that city.
City                     Monument
Delhi                    Red Fort
Agra                     Taj Mahal
Jaipur                   Jal Mahal


def display_monument(city):
    monuments = {
        "Delhi": "Red Fort",
        "Agra": "Taj Mahal",
        "Jaipur": "Jal Mahal"
    }

    return monuments.get(city, "Monument not found for the given city")

def main():
    city = input("Enter a city: ")
    monument = display_monument(city)
    print(f"The monument in {city} is: {monument}")

if __name__ == "__main__":
    main()
```

```python
4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

def count_divisions_by_three(number):
    count = 0
    while number > 10:
        number /= 3
        count += 1
    return count

def main():
    try:
        number = float(input("Enter a number: "))
        if number <= 10:
            print("Number should be greater than 10")
            return

        divisions = count_divisions_by_three(number)
        print(f"The given number can be divided by 3 {divisions} times before it becomes less than or equal to 10.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if __name__ == "__main__":
    main()
```

```python
5. Why and When to Use while Loop in Python give a detailed description with example

Why Use a while Loop:
1.Iterative Process: When you need to repeat a set of statements or actions until a specific condition is satisfied.
2.Dynamic Condition: When the number of iterations is not known beforehand and depends on some dynamic condition.
3.Indefinite Looping: When you want to keep iterating until a certain condition is met, without knowing the exact number of iterations in advance.

When to Use a while Loop:
1.Input Validation: When you need to repeatedly prompt the user for input until they provide a valid input.
2.Countdown or Countup: When you need to iterate a specific number of times in either direction (counting down or up).
3.Processing Events: When you need to continuously process events or perform tasks until a termination condition is reached.
4.Polling: When you need to continuously check for a certain condition or event and take action accordingly.
5.Simulation: When simulating real-world scenarios where events occur over time until certain conditions are met.

Example:
Let's consider an example where we want to repeatedly prompt the user to guess a secret number until they guess it correctly:

import random

# Generate a secret number
secret_number = random.randint(1, 100)

# Initialize guess count
guess_count = 0

# Start a while loop to repeatedly prompt the user for their guess
while True:
    guess = int(input("Guess the secret number (between 1 and 100): "))
    guess_count += 1
    
    if guess == secret_number:
        print(f"Congratulations! You guessed the secret number {secret_number} in {guess_count} attempts.")
        break  # Exit the loop if the guess is correct
    elif guess < secret_number:
        print("Try a higher number.")
    else:
        print("Try a lower number.")
```

```python
7. Reverse a while loop to display numbers from 10 to 1.

To display numbers from 10 to 1 using a while loop in reverse order, you can modify the loop control variables accordingly.

num = 10  # Start from 10

while num >= 1:  # Iterate while num is greater than or equal to 1
    print(num)
    num -= 1  # Decrement num in each iteration
```
